```python
#from imblearn.over_sampling import SMOTE
#x,y=SMOTE().fit_resample(x,y)
```

```python
df['text']=df['text'].str.lower()
df['text'] = df['text'].str.replace(r'[^\w\s]', '')  # Removes punctuation and special characters
df['text'] = df['text'].str.replace(r'\d+', '')     # Removes digits
df['text'] = df['text'].str.replace(r'\n', '')      # Removes newline characters
df['text'] = df['text'].str.replace(r'\r+', '')     # Removes carriage return characters
```

```python
x=df['text']
y=df['class']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#pipeline, yapılacak işlemleri sıraya koymak
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([('tfidf',TfidfVectorizer()),('classifier',RandomForestClassifier())])  # TfidfVectorizer() içine örneğin parametre olarak stopwords,ngram düzenlemeleri ekleyebiliriz
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y, test_size=.2,random_state=42)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.svm import SVC
```

```python
svm=Pipeline([('tfidf',TfidfVectorizer()),('classifier',SVC())])
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred=svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
print(classification_report(y_test,y_pred))
```
