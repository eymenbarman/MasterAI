# Data Imputation - Boş verileri doldurma

```python
# En etkili yöntemleden biri linear regression ile doldurma rakam dolduruyorsak
```

```python
import pandas as pd
```

```python
df=pd.read_csv("ExampleData.csv")
```

```python
df  #NaN-boş veriler var
```

```python
# mean ile, median ile, mode ile örneği doldurabiliriz.
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
imputed_data=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data)   #tekrar dataframe e çeviririz
```

```python
#en yakın komşuyla doldur benzerliklere göre doldurur-knn imputer
```

```python
from sklearn.impute import KNNImputer
```

```python
knn=KNNImputer()
```

```python
imputed_data2=knn.fit_transform(df)
```

```python
pd.DataFrame(imputed_data2)
```

```python
pd.DataFrame(imputed_data2,columns=[df.columns])  #sütun isimlerini geri getirdik
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='mean') #text lerde mode ile doldurulabilir örneğin
```

```python
imputed_data3=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data3, columns=[df.columns])
```
