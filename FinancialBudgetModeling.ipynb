```python
import pandas as pd

# Veri kümesini oku
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv"
df = pd.read_csv(url)

# İlk birkaç satıra bak
print(df.head())

# Veri tipi ve eksik veriler
print(df.info())
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# En fazla bütçe alan ilk 10 bakanlık
top10 = df.sort_values(by='Fund allotted(in ₹crores)', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top10['Department /Ministry'], top10['Fund allotted(in ₹crores)'])
plt.xlabel('Allocated Budget (₹ Crores)')
plt.title('Top 10 Ministries by Budget Allocation')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

```
