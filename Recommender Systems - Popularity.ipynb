```python
#en popüler film star wars gibi duruyor ilk bakışta şimdi bunları ölçelim matematiksel, kaç yorum yapılmış, ortalaması kaç, en yüksek rating'i alan film bunlardan
```

```python
movie_grouped=ratings.groupby('title').agg({'rating':[np.size,np.sum,np.mean]})
```

```python
movie_grouped
```

```python
popular_movies=movie_grouped.sort_values(('rating','mean'),ascending=False)
```

```python
popular_movies.head()
```

```python
grouped_sum=movie_grouped['rating']['sum'].sum()
```

```python
popular_movies['percentage']=movie_grouped['rating']['sum'].div(grouped_sum)*100
```

```python
popular_movies.head()
```

```python
popular_movies=popular_movies.sort_values(('percentage'),ascending=False)
```

```python
popular_movies
```

```python
popular_movies['Rank']=popular_movies['percentage'].rank(ascending=False)
```
